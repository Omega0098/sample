####  Standardization

```python
import pandas as pd
df = pd.read_csv('titanic.csv',usecols = ['Pclass','Age','Fare','Survived'])
df
```

```python
df.Age.fillna(df.Age.median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_sclaed = pd.DataFrame(df_scaled)
```

```python
import matplotlib.pyplot as plt
plt.hist(df_scaled[:,1],bins=20)
plt.hist(df['Pclass'],bins = 20)
```

#### Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
df_min_max = pd.DataFrame(min_max.fit_transform(df),columns = df.columns)
df_min_max.head()
```

```python
plt.hist(df_min_max['Pclass'],bins = 20,color='red')
plt.hist(df_min_max['Age'],bins = 20,color='blue')
plt.hist(df_min_max['Fare'],bins = 20,color='green')
```

```python
plt.hist(df_min_max['Age'],bins = 20)
```

```python
plt.hist(df_min_max['Fare'],bins = 20)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df_robust = pd.DataFrame(scaler.fit_transform(df),columns = df.columns)
df_robust.head()
```

```python
plt.hist(df_robust.Age,bins = 20)
```

```python
import scipy.stats as stat
import pylab
def  plot_data(df,feature):
    plt.figure(figsize = (10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
for i in df1.columns:
    plot_data(df1,i)
```

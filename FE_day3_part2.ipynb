```python
# Ordinal Encoding
import pandas as pd
import datetime as dte
```

```python
today = dte.datetime.today()
```

```python
today
```

```python
today-dte.timedelta(2)
```

```python
days = [today-dte.timedelta(x) for x in range(0,15)]
```

```python
days
```

```python
data = pd.DataFrame(days)
```

```python
data.columns = ['Days']
```

```python
data
```

```python
data['Weekday_ordinal']=(data['Days'].dt.weekday)+1
```

```python
data['Days'].dt.month
```

```python
dictionary = {
    1 : 'Monday',
    2 : 'Tuesday',
    3 : 'Wednesday',
    4 : 'Thrusday',
    5 : 'Friday',
    6 : 'Saturday',
    7 : 'Sunday'}
```

```python
data['Weekday'] = data['Weekday_ordinal'].map(dictionary)
```

```python
data
```

```python
##  Count or Frequency Encoding
train = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header = None,index_col = None)
```

```python
train
```

```python
train[1].unique()
```

```python
len(train[1].unique())
```

```python
columns = [1,3,5,6,7,8,9,13]
train = train[columns] 
```

```python
train.columns = ['Employemnet','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train.head()
```

```python
for feature in train.columns[:]:
    print(feature, ' : ', len(train[feature].unique()),' labels.')
```

```python
# Count Frequency Encoding
country_map = train['Country'].value_counts()
```

```python
country_map
```

```python
# Count Frequency Encoding
country_mapp = train['Country'].value_counts().to_dict()
```

```python
country_mapp
```

```python
train.Country = train.Country.map(country_mapp)
```
